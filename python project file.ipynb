```python
cone_type={1:['plain_cone',1.5], 2:["waffle_cone",2], 3:['cup',1]}
scoop_type={1:['vanilla',0.5], 2:['strawberry',0.5], 3:['chocolate',0.5], 4:['caramel',0.5], 5:['mint',0.5],6:['rainbow',0.5],7:['coffee',0.5],8:['bubble_gum',0.5]}
topping_choices={1:['peanuts',0.75],2:['caramel_sauce',0.75],3:['rainbow_sprinkles',0.5],4:['pecan',1],5:['chocolate_sprinkles',0.5]}
lst=[]
print('\033[1mWelcome to Shop\033[0m'.center(50))
name=input('enter name ')
for char in name:
    if char in ('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'):
         continue
        
    else:
        print('invalid name')
        break
       
    
        
print('Welcome mr.',name,',What would you like to take?')
n=int(input('how many ice cream you want  '))

for i in range(1,n+1):
    print('Ice cream number: ',i)
    print('\033[1mMENU\033[0m'.center(50))
    print('\033[1m Type\033[0m',end='                                   ')
    print('\033[1m Price\033[0m')
    print('1.plain cone',       end='                              ')
    print('1.5')
    print('2.waffle cone',      end='                              ')
    print('2')
    print('3.cup',              end='                                      ')
    print('1')
    
    price=0
    while True:
        ct=int(input('Select cone type '))
        if ct==1:
            price=price + cone_type[1][1]
            break
        if ct==2:
            price=price + cone_type[2][1]
            break
        if ct==3:
            price=price + cone_type[3][1]
            break
        else:
            continue
    print(f'Price={price}')
    
    sa=int(input('scoop amount '))
    
    while True:
        if sa==1 or sa==2 or sa==3:
            break
        else:
            print('Only upto 3 scoops available') 
            
    print('select scoop flavour'.center(50))
    
    print('\033[1mScoop Flavour\033[0m',end='                            ')
    print('\033[1mPrice\033[0m')
    print('1.Vanilla',end='                                 ')
    print('0.5')
    print('2.strawberry',end='                              ')
    print('0.5')
    print('3.chocolate',end='                               ')
    print('0.5')
    print('4.caramel',end='                                 ')
    print('0.5')
    print('5.Mint',end='                                    ')
    print('0.5')
    print('6.Rainbow',end='                                 ')
    print('0.5')
    print('7.Coffee',end='                                  ')
    print('0.5')
    print('8.Bubble gum',end='                              ')
    print('0.5')
    
    st=int(input('enter scoop flavour number'))
    print('scoop flavour selected:',scoop_type[st][0])
    
    while True:
        
        if sa==1:
            price=price+0.5*sa
            break
        elif sa==2:
            price=price+0.5*sa
            break
        elif sa==3:
            price=price+0.5*sa
            break
        else:
            continue
    print(f'Price={price}')
    
    ta=int(input('Topping amount'))
    while True:
        if ta==1 or ta==2 or ta==3 or ta==4:
            break
        else:
            print('only upto 4 toppings availabe')
            continue
    
    print('select Topping '.center(50))
    print('\033[1mToppings\033[0m',end='                                 ')
    print('\033[1mPrice\033[0m')
    print('1.Peanuts',end='                                ')
    print('0.75')
    print('2.Caramel Sauce',end='                          ')
    print('0.5')
    print('3.Rainbow Sprinkles',end='                      ')
    print('0.5')
    print('4.Pecan',end='                                  ')
    print('1')
    print('5.Chocolate Sprinkles')
    
    tf=int(input('enter topping number '))
    
    while True:
        if tf==1:
            price=price+topping_choices[1][1]*ta
            break
        elif tf==2:
            price=price+topping_choices[2][1]*ta
            break
        elif tf==3:
            price=price+topping_choices[3][1]*ta
            break
        elif tf==4:
            price=price+topping_choices[4][1]*ta
            break
        elif tf==5:
            price=price+topping_choices[5][1]*ta
            break
        else:
            print('wrong input')
            continue
    lst.append(price)
    print('cost: ',price)
print('Total Bill: ',sum(lst))
    
```
